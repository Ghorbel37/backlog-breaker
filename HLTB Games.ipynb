```python
import os
import csv

# Define the folder containing the shortcuts
folder_path = r"Path\To\Your\Shortcuts\Folder"

# Define the output CSV file path
output_file_path = r"Shortcuts.csv"

# Create a list to store shortcut names
shortcut_names = []

# Get all shortcuts (.lnk and .url) in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".lnk") or file_name.endswith(".url"):
        shortcut_names.append(os.path.splitext(file_name)[0])

# Write shortcut names to a CSV file
with open(output_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["ShortcutName"])
    writer.writerows([[name] for name in shortcut_names])
```

```python
from howlongtobeatpy import HowLongToBeat
import pandas as pd

# Replace 'file_path.csv' with the path to your CSV file
file_path = r"Shortcuts.csv"

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

# Create an instance of HowLongToBeat
hltb = HowLongToBeat()

# Create new columns to store search results
df['GameName'] = ""
df['ReleaseYear'] = ""
df['MainStory'] = ""
df['MainExtra'] = ""
df['Completionist'] = ""
df['AllStyles'] = ""
# df["Genre"] = ""

# Now you can continue with your existing code
for index, shortcut_name in enumerate(df['ShortcutName']):
    # Perform the search using HowLongToBeat
    results = hltb.search(shortcut_name)
    
    if results is not None and len(results) > 0:
        # Filter out DLC entries and get the best matching full game entry
        full_game_results = [entry for entry in results if entry.game_type != 'dlc']
        if full_game_results:
            best_element = max(full_game_results, key=lambda element: element.similarity)
            # Store search results in DataFrame
            df.at[index, 'GameName'] = best_element.game_name
            df.at[index, 'ReleaseYear'] = best_element.release_world
            df.at[index, 'MainStory'] = best_element.main_story
            df.at[index, 'MainExtra'] = best_element.main_extra
            df.at[index, 'Completionist'] = best_element.completionist
            df.at[index, 'AllStyles'] = best_element.all_styles
            # df.at[index, 'Genre'] = best_element.json_content
            # Print the main story completion time for the best matching full game entry
            print(f"Search results for '{best_element.game_name} {best_element.release_world}': {best_element.main_story},{best_element.main_extra},{best_element.completionist},{best_element.all_styles}")
        else:
            print(f"No full game found for '{shortcut_name}'")
    else:
        print(f"No results found for '{shortcut_name}'")


df.drop(columns=["ShortcutName"], inplace=True)
df.to_excel('modified_file.xlsx', index=True)
```
